```python


cat_cols = ['Sex', 'Embarked', 'Survived', 'Pclass']

# Create subplots: 1 row, 4 columns
fig, axes = plt.subplots(1, len(cat_cols), figsize=(12, 3))

for i, col in enumerate(cat_cols):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(f"{col} Count")
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python

num_cols = ['Age', 'Fare', 'SibSp', 'Parch']

fig, axes = plt.subplots(1, len(num_cols), figsize=(12, 3))

for i, col in enumerate(num_cols):
    sns.histplot(df[col], kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f"{col} Dist")
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python


# Select only relevant numeric columns
num_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

plt.figure(figsize=(8,5))
corr_matrix = df[num_cols].corr()  # correlation only for selected columns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Titanic Dataset - Correlation Heatmap")
plt.tight_layout()
plt.show()
```
